# Part I. Image retrieval and visualization.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

The following part loads the image records file which is stored in npy format and then transforms it into a dictionary. For each image data, it corresponds to several boxes that enclose the cat in the image. 

```python
# get all the keys 
annotation = np.load("DATA/annotation_list.npy")[()] # annotation originally is a ndarray, the [()] transform it into a dict
print(annotation)
type(annotation)  
```

```python
image_list = annotation.keys()   # Retrieve the keys associated with the image list
image_list
len(image_list)
image_list_train = annotation.keys()[0:448]
len(image_list_train)
```

Next we visualize an image containing cats and boxes enclosing them. 

```python
# get a viusalization 
imgfile = "2008_006817"
img = plt.imread("DATA/{}.jpg".format(imgfile))     # read the image using plt
box = annotation[imgfile]     # retrieve the detection boxes
print box
# Create figure and axes
fig,ax = plt.subplots(1)     # generate a canvas 
ax.imshow(img)
for eachbox in box:
    cx, cy, w, h = eachbox 

    # Create a Rectangle patch
    rect = patches.Rectangle((cx-w/2,cy-h/2),w,h,linewidth=1,edgecolor='y',facecolor='none')
    # Draw a rectangle with lower left at xy = (x, y) with specified width and height
    # Add the patch to the Axes
    ax.add_patch(rect)

plt.show()

print "The box is {}".format(box)
print "image size is {}".format(img.shape)
```
